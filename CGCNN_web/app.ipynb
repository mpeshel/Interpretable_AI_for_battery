```python

from flask import Flask, flash, render_template, url_for, request, redirect
from id_voltage import id_get_result
from file_voltage import file_get_result
from file_voltage import allowed_file
from werkzeug.utils import secure_filename
import os

app = Flask(__name__)


@app.route('/', methods=['POST', 'GET'])

def index():
    if request.method == 'POST':
        material_id = request.form['material_id']
        working_ion = request.form['working_ion']
        cif_file=request.files['file']
        if cif_file:
            
            if allowed_file(cif_file.filename):
                file = secure_filename(cif_file.filename)
                cif_file.save(file)
                return redirect(f'/upload_file/{file}/{working_ion}')
                
        elif material_id:
            return redirect(f'/submit_id/{material_id}/{working_ion}')
        else:
            return redirect('/invalid_id')
    else:
        return render_template('index.html')

    
@app.route('/upload_file/<file>/<ion>', methods=['GET', 'POST'])
def upload_file(file, ion):
    if request.method == 'POST':
        return redirect('/')
    else:
        try:
            result = file_get_result(file, ion)
            os.remove(file)
            return render_template('upload.html', result=result)
        except:
            return redirect('/invalid_id')


@app.route('/submit_id/<id>/<ion>', methods=['GET', 'POST'])
def submit_id(id, ion):
    if request.method == 'POST':
        return redirect('/')
    else:
        try:
            result = id_get_result(id, ion)
            return render_template('predict.html', result=result)
        except:
            return redirect('/invalid_id')

@app.route('/invalid_id', methods=['GET', 'POST'])
def invalid_id():
    if request.method == 'POST':
        return redirect('/')
    else:
        return render_template('invalid.html')

@app.route('/about', methods=['GET', 'POST'])
def about():
    if request.method == 'POST':
        return redirect('/')
    else:
        return render_template('about.html')

if __name__ == "__main__":
    app.run(debug=True, use_reloader=False)
```
